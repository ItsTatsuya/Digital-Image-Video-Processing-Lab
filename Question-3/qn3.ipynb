```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = '../input.jpg'
img = cv2.imread(img_path)
if img is None:
    raise FileNotFoundError(f"Cannot load image at {img_path}")
# prepare a grayscale version
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

equalized = cv2.equalizeHist(gray)

# display side-by-side
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].imshow(gray, cmap='gray')
axes[0].set_title("Original Grayscale")
axes[0].axis('off')
axes[1].imshow(equalized, cmap='gray')
axes[1].set_title("Equalized Grayscale")
axes[1].axis('off')
plt.show()

# also plot histograms for before/after
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].hist(gray.ravel(), bins=256, range=(0,255), color='black')
axes[0].set_title("Original Hist")
axes[1].hist(equalized.ravel(), bins=256, range=(0,255), color='black')
axes[1].set_title("Equalized Hist")
plt.show()
```

```python
# convert to YCrCb
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
y, cr, cb = cv2.split(ycrcb)
y_eq = cv2.equalizeHist(y)
# merge back and convert to RGB for display
ycrcb_eq = cv2.merge([y_eq, cr, cb])
bgr_eq = cv2.cvtColor(ycrcb_eq, cv2.COLOR_YCrCb2BGR)
rgb_eq = cv2.cvtColor(bgr_eq, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(rgb_eq)
plt.title("Color Image Equalized (Y‑Cr‑Cb)")
plt.axis('off')
```
