```python

# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = os.path.join('..', '..', 'input_noisy.jpg')

original_image = cv2.imread(image_path)

# Check if the image was loaded successfully
if original_image is None:
    print(f"Error: Could not open or find the image at '{image_path}'")
else:
    # Convert the image from BGR to RGB for displaying with matplotlib
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Define the kernel sizes for averaging
    kernel_sizes = [2, 8, 16, 32, 128]

    # Create a list to store the averaged images
    averaged_images = []

    # Perform spatial averaging (blurring) for each kernel size
    for k_size in kernel_sizes:
        # cv2.blur performs averaging with a kernel of specified size
        # A k_size x k_size kernel is used.
        averaged_img = cv2.blur(original_image_rgb, (k_size, k_size))
        averaged_images.append(averaged_img)

    # Display the results for comparison
    # Create a figure to display all the images
    plt.figure(figsize=(20, 10))

    # Display the original noisy image
    plt.subplot(2, 3, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Noisy Image')
    plt.axis('off')

    # Display the averaged images
    for i, k_size in enumerate(kernel_sizes):
        plt.subplot(2, 3, i + 2)
        plt.imshow(averaged_images[i])
        plt.title(f'Averaged with {k_size}x{k_size} kernel')
        plt.axis('off')

    # Show the plot
    plt.tight_layout()
    plt.show()

    # Comment on the results
    # By observing the output images, we can see that as the kernel size for averaging increases,
    # the image becomes smoother and the noise is reduced. However, a very large kernel size
    # can also lead to a loss of detail, making the image blurry.

    # best result is found at kernel size 2
```
