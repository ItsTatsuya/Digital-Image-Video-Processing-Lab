```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a black image
image1 = np.zeros((300, 300), dtype="uint8")
# Draw a white rectangle on it
cv2.rectangle(image1, (50, 50), (250, 250), 255, -1)
cv2.imwrite('image1.png', image1)

# Create another black image
image2 = np.zeros((300, 300), dtype="uint8")
# Draw a white circle on it
cv2.circle(image2, (150, 150), 100, 255, -1)
cv2.imwrite('image2.png', image2)

# Load the images
img1 = cv2.imread('image1.png', 0)
img2 = cv2.imread('image2.png', 0)

# Display the original images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Image 1")
plt.imshow(img1, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Image 2")
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
add_img = cv2.add(img1, img2)
plt.imshow(add_img, cmap='gray')
plt.title("Added Image")
plt.show()
```

```python
sub_img = cv2.subtract(img1, img2)
plt.imshow(sub_img, cmap='gray')
plt.title("Subtracted Image")
plt.show()
```

```python
mul_img = cv2.multiply(img1, img2)
plt.imshow(mul_img, cmap='gray')
plt.title("Multiplied Image")
plt.show()
```

```python
# Convert images to float32 for division
img1_float = img1.astype(np.float32)
img2_float = img2.astype(np.float32)

div_img = cv2.divide(img1_float, img2_float + 1e-5)
plt.imshow(div_img, cmap='gray')
plt.title("Divided Image")
plt.show()
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(3, 2, 1)
plt.title("Image 1")
plt.imshow(img1, cmap='gray')

plt.subplot(3, 2, 2)
plt.title("Image 2")
plt.imshow(img2, cmap='gray')

plt.subplot(3, 2, 3)
plt.title("Added")
plt.imshow(add_img, cmap='gray')

plt.subplot(3, 2, 4)
plt.title("Subtracted")
plt.imshow(sub_img, cmap='gray')

plt.subplot(3, 2, 5)
plt.title("Multiplied")
plt.imshow(mul_img, cmap='gray')

plt.subplot(3, 2, 6)
plt.title("Divided")
plt.imshow(div_img, cmap='gray')

plt.tight_layout()
plt.show()
```
