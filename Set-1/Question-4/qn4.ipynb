```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = '../input.jpg'
img = cv2.imread(img_path)
if img is None:
    raise FileNotFoundError(f"Cannot load image at {img_path}")

# Convert to grayscale for local equalization
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Create CLAHE object: clipLimit=2.0, tileGridSize=8×8
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
local_eq = clahe.apply(gray)

# Display side‑by‑side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(gray, cmap='gray')
ax1.set_title("Original Grayscale")
ax1.axis('off')
ax2.imshow(local_eq, cmap='gray')
ax2.set_title("Local Equalization (CLAHE)")
ax2.axis('off')
plt.show()
```

```python
fig, (h1, h2) = plt.subplots(1, 2, figsize=(10,4))
h1.hist(gray.ravel(), bins=256, range=(0,255), color='gray')
h1.set_title("Global Hist")
h2.hist(local_eq.ravel(), bins=256, range=(0,255), color='gray')
h2.set_title("Local (CLAHE) Hist")
plt.show()
```
